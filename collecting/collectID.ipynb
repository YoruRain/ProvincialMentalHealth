```python
import os

id_list_dir = r"weibo_user_id"

id_list_texts = [text for text in os.listdir(id_list_dir) if len(text) < 10]

id_list = []

for text in id_list_texts:
    with open(os.path.join(id_list_dir, text), "r") as f:
        ids = f.readlines()
        for id in ids:
            id_list.append(id.strip())

print(len(id_list))

id_set = set(id_list)

print(len(id_set))
```

```python
import os

id_list_dir = r"weibo_user_id"

id_ip_list_texts = [text for text in os.listdir(id_list_dir) if len(text) > 10]

id_ip_list = []

for text in id_ip_list_texts:
    with open(os.path.join(id_list_dir, text), "r") as f:
        id_ips = f.readlines()
        for id_ip in id_ips:
            id, ip = id_ip.strip().split(',')
            id_ip_list.append((id, ip))

print(len(id_ip_list))

id_ip_set = set(id_ip_list)

print(len(id_ip_set))
```

```python
id_remain = [id_ip[0] for id_ip in id_ip_set]
id_removed = [uid for uid in id_set if uid not in id_remain]

with open("user_id_removed.txt", 'w', encoding='utf-8') as f:
    f.writelines(f"{uid}\n" for uid in id_removed)
```

```python
id_before = []
id_file_before = r"collecting1\user_id.txt"
with open(id_file_before, 'r', encoding='utf-8') as f:
    id_before = f.readlines()

id_before = [id.strip() for id in id_before]

id_ip_set = {id_ip for id_ip in id_ip_set if id_ip[0] not in id_before}

len(id_ip_set)
```

```python
def remove_non_mainland(id_ips):
    provinces = ['北京', '天津', '河北', '山西', '内蒙古', '辽宁', '吉林', '黑龙江', 
             '上海', '江苏', '浙江', '安徽', '福建', '江西', '山东', '河南', 
             '湖北', '湖南', '广东', '广西', '海南', '重庆', '四川', '贵州', 
             '云南', '西藏', '陕西', '甘肃', '青海', '宁夏', '新疆']
    
    users_to_remove = [id_ip for id_ip in id_ips if id_ip[1] not in provinces]
    
    for user in users_to_remove:
        id_ips.remove(user)
        
    
remove_non_mainland(id_ip_set)

len(id_ip_set)
```

```python
with open(r"user_id_removed_cleaned_with_ip.txt", "r", encoding="utf-8") as f:
    id_ip_set = {tuple(line.strip().split(",")) for line in f}

remove_non_mainland(id_ip_set)

for id_ip in id_ip_set:
    print(f"{id_ip[0]}")
```

```python
with open("user_id_ip.txt", 'w', encoding='utf-8') as f:
    f.writelines(f"{id_ip[0]},{id_ip[1]}\n" for id_ip in id_ip_set)
```
